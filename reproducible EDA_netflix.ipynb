```python
import pandas as pd
```

```python
df = pd.read_csv("netflix_titles.csv")
df
```

```python
df.replace("", pd.NA, inplace = True)
df.head()
```

```python
df_clean = df.copy()
df_clean
```

```python
# 1. Clean duration for MOVIES
df_clean["duration_min"] = df_clean["duration"].str.replace(" min", "").where(df_clean["type"] == "Movie")
df_clean["duration_min"] = pd.to_numeric(df_clean["duration_min"], errors="coerce")

# 2. Clean duration for TV SHOWS
df_clean["duration_seasons"] = df_clean["duration"].str.replace(" Seasons", "").str.replace(" Season", "").where(df_clean["type"] == "TV Show")
df_clean["duration_seasons"] = pd.to_numeric(df_clean["duration_seasons"], errors="coerce")

# 3. Extract year from date_added: "24-Sep-21" → 2021
df_clean["year_added"] = pd.to_datetime(df_clean["date_added"], errors="coerce").dt.year

# Check your work!
df_clean[["type", "duration", "duration_min", "duration_seasons", "year_added"]].head(10)
```

```python
df_clean['duration_min'].isna().sum() #The total NaN in this column is 2679 - the same as in R
df_clean['duration_seasons'].isna().sum() #The total NaN in this column is 6131 - the same as in R
df_clean['year_added'].isna().sum() #The total NaN in this column is 98 - the same as in R
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Count how many movies vs tv shows
type_counts = df_clean["type"].value_counts()

# Plot it
plt.figure(figsize=(6, 4))
sns.countplot(data=df_clean, x="type", color="darkblue")

# Labels (same as xlab, ylab, ggtitle in R)
plt.title("Netflix Content by Type")
plt.xlabel("Type")
plt.ylabel("Count")

plt.show()
```

```python
df_rating = df_clean.dropna(subset=["rating"])

# Plot it
plt.figure(figsize=(10, 5))
sns.countplot(data=df_rating, x="rating", color="darkblue",
              order=df_rating["rating"].value_counts().index)

plt.title("Netflix Content by Rating")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(rotation=45)  # tilts the x labels so they don't overlap

plt.show()
```

```python
df_movies = df_clean[df_clean["type"] == "Movie"].copy()
```

```python
plt.figure(figsize=(16, 6))

ax = sns.histplot(data=df_movies, x="duration_min", hue="rating",
             binwidth=3, multiple="stack",
             palette="tab20")

plt.title("Movie Duration Distribution")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Count")

# Get the legend that seaborn already created and resize it
legend = ax.get_legend()                    # grab seaborn's own legend
legend.set_title("Rating")                  # set the title
legend.set_bbox_to_anchor((1.02, 1))        # move it to the right

# Make text bigger
for text in legend.get_texts():
    text.set_fontsize(11)                   # bigger rating labels
legend.get_title().set_fontsize(13)         # bigger title

plt.tight_layout()
plt.show()
```

```python
# Filter only 1 Season shows and drop NaN ratings
df_1season = df_clean[df_clean["duration"] == "1 Season"].dropna(subset=["rating"])

# Count combinations of rating and duration (same as count() in R)
season_counts = df_1season.groupby(["rating", "duration"]).size().reset_index(name="n")

# Pivot the table so it's ready for a heatmap
season_pivot = season_counts.pivot(index="rating", columns="duration", values="n")

# Plot it
plt.figure(figsize=(6, 8))
sns.heatmap(season_pivot, annot=True, fmt="g", cmap="Blues", linewidths=0.5)

plt.title("Rating of 1 Season TV Shows")
plt.xlabel("Duration")
plt.ylabel("Rating")

plt.tight_layout()
plt.show()
```

```python
# Count movies added per year
movies_per_year = df_clean[df_clean["type"] == "Movie"].dropna(subset=["year_added"])
movies_per_year = movies_per_year.groupby("year_added").size().reset_index(name="count")

# Count TV shows added per year
shows_per_year = df_clean[df_clean["type"] == "TV Show"].dropna(subset=["year_added"])
shows_per_year = shows_per_year.groupby("year_added").size().reset_index(name="count")

# Plot both lines on the same chart
plt.figure(figsize=(12, 6))

sns.lineplot(data=movies_per_year, x="year_added", y="count", 
             marker="o", label="Movies")
sns.lineplot(data=shows_per_year, x="year_added", y="count", 
             marker="o", label="TV Shows")

plt.title("Titles Added to Netflix Per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(range(int(df_clean["year_added"].min()), 
                 int(df_clean["year_added"].max())+1), rotation=45)
plt.legend(title="Type")

plt.tight_layout()
plt.show()
```

```python
df_box = df_clean[df_clean["type"] == "Movie"].dropna(subset=["rating", "duration_min"])

plt.figure(figsize=(14, 6))
sns.boxplot(data=df_box, x="rating", y="duration_min",
            hue="rating", legend=False,   # ← just these two changes
            palette="tab20")

plt.title("Movie Duration Distribution by Rating")
plt.xlabel("Rating")
plt.ylabel("Duration (Minutes)")
plt.xticks(rotation=45)
plt.yticks(range(0, 800, 100))

plt.tight_layout()
plt.show()
```
